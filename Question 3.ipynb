Consider the _company_sales.csv_ dataset.

```python
import pandas as pd
file_path = 'C:/STK/Assignment 1/company_sales.csv'
df = pd.read_csv(file_path)
```

An investor would like to decide on a company to invest in. The _company_sales.csv_ dataset contains the sales (in millions) over a selected time period for the 3 companies under consideration. 

1. Retrieve the first 5 rows of the data to see how it is structured.

```python
first_five_rows = df.head(5)
print("First five rows of Company Sales:")
print(first_five_rows)
```

```python
# printing column names
column_names = df.columns
print("Column names:")
for column in column_names:
    print(column)
```

2. Plot a line graph showing the sales over a time for the 3 companies in a single plot. Remember to label and format the graph appropriately to create an effective and meaningful visualisation. 

Some specifications:
- Include a legend in the top left corner.
- Use a black dashed line for the Nexus Innovations data
- Use a red solid line for the Quantum Industries data
- Use an orange dash-dot line for the StellarTech Solutions data

```python
import matplotlib.pyplot as plt

# Creating a line plot
plt.figure(figsize=(10, 6))
plt.plot(df['Period'], df['Nexus Innovations'], linestyle='--', color='black', label='Nexus Innovations')
plt.plot(df['Period'], df['Quantum Industries'], linestyle='-', color='red', label='Quantum Industries')
plt.plot(df['Period'], df['StellarTech Solutions'], linestyle='-.', color='orange', label='StellarTech Solutions')

# Adding labels and title
plt.xlabel('Period')
plt.ylabel('Sales')
plt.title('Sales for Three Companies Over Time')

# Adding a legend
plt.legend(loc='upper left')

# Printing the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

3. Based on the plot, which company would you recommend to the investor to invest in and why?

I would recommend StellarTech Solution as it shows higher growth rate in sales over time.
